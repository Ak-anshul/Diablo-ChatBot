```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def chatbot(User_Input):
    input = tokenizer.encode(User_Input + tokenizer.eos_token, return_tensors='pt')
    output = model.generate(input, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    return tokenizer.decode(output[:, input.shape[-1]:][0], skip_special_tokens=True)
```

```python
def main():
  print("Chatbot: Hello! I'm your DialoGPT-based chatbot. You can start chatting with me. Type 'quit' to exit.")
  while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            break
        response = chatbot(user_input)
        print("Chatbot:", response)
```

```python
main()
```
